```python
# Write a function definition named sum_all that takes in sequence of numbers and returns all the numbers added together.
def sum_all(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

sum_all([1, 2, 3])
```

```python
def factorial(n):
  "compute the factorial of a number n"
  result = 1
  for i in range(1,n+1):
    result = result*i
  return result
factorial(5)
```
